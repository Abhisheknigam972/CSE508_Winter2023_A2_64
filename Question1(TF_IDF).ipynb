```python
#Unique Words
folder_path = "/content/drive/MyDrive/Mtech/Sem2/IR/My_assignment/Assignment2/CSE508_Winter2023_Dataset"

# Create an empty set to store the unique words
unique_words = set()

# Loop over each file in the folder
for file_name in os.listdir(folder_path):
        # Open the file
        with open(os.path.join(folder_path, file_name), "r") as f:
            # Read the file contents and split them into words
            words = set(f.read().split())

        # Add the unique words to the set
        unique_words |= words

# Print the unique words
print(unique_words)
```

```python
#Posting List

# Create an empty dictionary to store the posting list
posting_list = {}

# Loop over each file in the folder
for file_name in os.listdir(folder_path):
        # Open the file
        with open(os.path.join(folder_path, file_name), "r") as f:
            # Read the file contents and split them into words
            words = f.read().split()

        # Loop over each word and update the posting list
        for word in set(words):
            if word in posting_list:
                posting_list[word].add(file_name)
            else:
                posting_list[word] = {file_name}

# Print the posting list
for word, files in posting_list.items():
    print(word, files)
```

```python
import math

num_docs = len(os.listdir(folder_path))
idf_values = {}
for term, files in posting_list.items():
    doc_freq = len(files)
    idf_values[term] = math.log2((num_docs) / (doc_freq+1))

# Print the IDF values
for term, idf in idf_values.items():
    print(term, idf)
```

**Term Frequency Weighting Scheme**

```python
import pandas as pd
dtermfreq = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique_words)
```

```python
def termfreq(folder_path):
    for i, file in enumerate(os.listdir(folder_path)):
        with open(os.path.join(folder_path, file), "r") as f:
            words = f.read().split()
            unique_words = set(words)
            word_freq = {}
            max_freq = 1  # default value to avoid division by zero
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
                max_freq = max(max_freq, word_freq[word])
            for word in word_freq:
                tf = word_freq[word] / max_freq
                idf = idf_values.get(word, 1)
                tf_idf = tf * idf
                print(word,tf_idf)
                dtermfreq.loc[file, word] = tf_idf
    return dtermfreq
```

**Double Normalization Weighting Scheme:**

```python
import pandas as pd
ddoublenorm = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique_words)
```

```python
def doublenorm(folder_path):
    for i, file in enumerate(os.listdir(folder_path)):
        with open(os.path.join(folder_path, file), "r") as f:
            words = f.read().split()
            unique_words = set(words)
            word_freq = {}
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
            max_freq = max(word_freq.values())
            for word in word_freq:
                tf = 0.5 + 0.5 * (word_freq[word] / max_freq)
                idf = idf_values.get(word, 1)
                tf_idf = tf * idf
                print(word,tf_idf)
                ddoublenorm.loc[file, word] = tf_idf
    return ddoublenorm
```
